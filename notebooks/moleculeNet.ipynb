```python
from pathlib import Path as pt
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import sigfig
import json
import uncertainties as uc
import numpy as np
from sklearn import metrics
```

```python
# ESOL: Water solubility data(log solubility in mols per litre) for common organic small molecules.
esol = '/Users/aravindhnivas/Documents/MoleculeNet/delaney-processed.csv'
df = pd.read_csv(esol)
df
```

```python
df.columns
```

```python
y_true = df['measured log solubility in mols per litre']
y_pred = df['ESOL predicted log solubility in mols per litre']

r2 = metrics.r2_score(y_true, y_pred)
mae = metrics.mean_absolute_error(y_true, y_pred)
mse = metrics.mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
print(f'R^2: {r2:.2f}')
print(f'MAE: {mae:.2f}')
print(f'MSE: {mse:.2f}')
print(f'RMSE: {rmse:.2f}')
```
